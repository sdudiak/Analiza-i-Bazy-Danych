```python
# Iports
import numpy as np
import pandas as pd
import re
```

```python
# Read the original data
original_data = pd.read_csv("../OriginalData/tb.csv")
display(original_data)
```

```python
# Get rid of NaN values
data = original_data.dropna()
# remove new_sp column
data = data.drop(columns=['new_sp'])
data = data.drop(columns=['new_sp_m014'])
data = data.drop(columns=['new_sp_f014'])
# remove 'new_sp' from column names
data.columns = data.columns.str.replace("new_sp_", " ")
display(data)
# devide dataframe into two dataframes
data_males = data.filter(regex='m')
data_males = data_males.join(data['year'])
data_males = data_males.join(data['iso2'])

data_females = data.filter(regex='f')
data_females = data_females.join(data['year'])
data_females = data_females.join(data['iso2'])

display(data_males)
display(data_females)
```

```python
# Group data by the year to get global cases for each year
data_males_global = data_males.groupby(['year']).sum()
data_females_global = data_females.groupby(['year']).sum()

#Rename columns
data_males_global.columns = ['0-4', '5-14','15-24','25-34','35-44','45-54','55-64','65+','unasigned']
data_females_global.columns = ['0-4', '5-14','15-24','25-34','35-44','45-54','55-64','65+','unasigned']
data_males_global = data_males_global.astype(int)
data_females_global = data_males_global.astype(int)

display(data_males_global)
display(data_males_global)

# Save prepared data
data_males_global.to_csv("../AnalysisData/global_tuberculosis_cases_2005-2008_males.csv")
data_females_global.to_csv("../AnalysisData/global_tuberculosis_cases_2005-2008_females.csv")
```
