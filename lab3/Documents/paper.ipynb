```python
import pandas as pd
import matplotlib.pyplot as plt
data_males = pd.read_csv("../AnalysisData/global_tuberculosis_cases_2005-2008_males.csv")
data_females = pd.read_csv("../AnalysisData/global_tuberculosis_cases_2005-2008_females.csv")




plt.figure()
vals = list(data_males.iloc[0][1:].values)
keys = list(data_males.iloc[0][1:].keys())
plt.title("Global male tuberculosis cases in 2005")
plt.xlabel("age")
plt.ylabel("number of cases")
plt.bar(keys,vals)
plt.show()

plt.figure()
vals = list(data_females.iloc[0][1:].values)
keys = list(data_females.iloc[0][1:].keys())
plt.title("Global female tuberculosis cases in 2005")
plt.xlabel("Age")
plt.ylabel("Number of cases")
plt.bar(keys,vals)
plt.show()
```
